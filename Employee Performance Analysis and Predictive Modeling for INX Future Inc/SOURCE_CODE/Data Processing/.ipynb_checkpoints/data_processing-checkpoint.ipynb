# Preprocessing 

## Importing Necessary Libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler,MinMaxScaler,OrdinalEncoder
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
df = pd.read_excel('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls')
pd.set_option('display.max_columns',None)
```

## Encoding Categorical Columns

### One-Hot Encoding

```python
ohe_cols = ['Attrition','Gender']
ohe = OneHotEncoder(drop = "first",sparse_output=False)
ohe_result = ohe.fit_transform(df[ohe_cols])
df[ohe_cols] = pd.DataFrame(ohe_result)
df[ohe_cols].head()
```

### Label Encoding

```python
le_cols = ['OverTime','EmpJobRole','EmpDepartment','EducationBackground','MaritalStatus']
le_encoders = {}
for i in le_cols:
    le = LabelEncoder()
    df[i] = le.fit_transform(df[i])
    le_encoders[i] = le
df[le_cols].head()
```

```python
le_encoders
```

### Ordinal Encoding

```python
ord_cols = ['BusinessTravelFrequency','PerformanceRating']
ord_encoders = {}
for i in ord_cols:
    ord = OrdinalEncoder()
    df[[i]] = ord.fit_transform(df[[i]])
    ord_encoders[i] = ord
```

```python
ord_encoders
```

```python
df[ord_cols].head()
```

## correlation Analysis

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.drop(columns="EmpNumber").corr(),cmap="Blues")
```

```python
df.drop(columns="EmpNumber", axis=1, inplace=True)
```

```python
pd.DataFrame(df.corr().PerformanceRating.sort_values(ascending = False)[:4])
```

## Saving Processed Data

```python
df.to_csv("processed_data.csv", index= False)
```

## Packing Encoders

```python
with open ("one_hot_enc", "wb") as file:
    pickle.dump(ohe,file)
```

```python
with open ("label_encs", "wb") as file:
    pickle.dump(le_encoders,file)
```

```python
with open ("ordinal_encs", "wb") as file:
    pickle.dump(ord_encoders,file)
```
