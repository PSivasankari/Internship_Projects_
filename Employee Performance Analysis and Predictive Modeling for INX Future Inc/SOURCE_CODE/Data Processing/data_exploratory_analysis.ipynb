## Univariate Analysis

```python
cat_cols = df.drop(columns =['EmpNumber','EmpJobRole']).select_dtypes(include="O").columns
len(cat_cols)
```

```python
plt.figure(figsize=(18,20))
w = 1
for i in cat_cols:
    plt.subplot(4,2,w)
    plt.pie(x = df[i].value_counts(),
            autopct = "%1.1f%%",
            pctdistance= 1.2,
            textprops={'fontsize': 13} )
    plt.legend(df[i].value_counts().index,
               loc = 9, title = i,
               title_fontsize = 15,
               bbox_to_anchor=(1.4, 1),
               fontsize = 14)
    plt.title(f"Distribution of {i}",fontsize = 18)
    w = w+1
plt.tight_layout()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = df.EmpJobRole.value_counts().values,
            y = df.EmpJobRole.value_counts().index,
            palette="deep" )
plt.yticks(fontsize = 13)
plt.xticks(fontsize = 13)
plt.ylabel(ylabel="Job Roles",fontsize = 15)
plt.xlabel(xlabel="Count",fontsize = 15)
plt.show()
```

```python
for i in df.select_dtypes(exclude="O").columns:
    if len(df[i].value_counts()) <= 10:
        print(i,len(df[i].value_counts()))
```

```python
disc_cols = ["EmpEducationLevel","EmpEnvironmentSatisfaction","EmpJobInvolvement","EmpJobLevel","EmpJobSatisfaction","NumCompaniesWorked",
            "EmpRelationshipSatisfaction","TrainingTimesLastYear","EmpWorkLifeBalance","PerformanceRating"]
len(disc_cols)
```

```python
plt.figure(figsize=(15, 20))
w = 1
for i in disc_cols:
    plt.subplot(5, 2, w)
    sns.barplot(
        x=df[i].value_counts().index,    
        y=df[i].value_counts().values,   
        palette="deep")
    plt.ylabel("Count", fontsize=12)   
    plt.xlabel(i, fontsize=12) 
    plt.title(f"Distribution of {i}", fontsize=14)
    w = w + 1
plt.tight_layout()
```

```python
for i in df.select_dtypes(exclude="O").columns:
    if len(df[i].value_counts()) > 10:
        print(i,len(df[i].value_counts()))
```

```python
cont_cols = ['Age','DistanceFromHome','EmpHourlyRate', 'EmpLastSalaryHikePercent','TotalWorkExperienceInYears', 
             'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
len(cont_cols)
```

```python
plt.figure(figsize=(15,15))
w = 1
for i in cont_cols:
    plt.subplot(3,3,w)
    sns.histplot(x = i, data = df, kde= True)
    w = w+1
    plt.title(f"Distribution of {i}", fontsize= 15)
    plt.xlabel(i,fontsize = 12)
plt.tight_layout()
```

## Bivariate Analysis

```python
cols = ["TotalWorkExperienceInYears","ExperienceYearsAtThisCompany", "EmpJobLevel","PerformanceRating",]
plt.figure(figsize=(17,10))
w = 1
for i in cols:
    plt.subplot(2,2,w)
    sns.lineplot(x = "Age", y = i , data = df)
    plt.title(f"Relationship Between Age and {i}", fontsize = 15)
    plt.xlabel("Age",fontsize = 13)
    plt.ylabel(i,fontsize = 13)
    w = w+1
plt.tight_layout()
```

```python
cols = ["EmpLastSalaryHikePercent","EmpEnvironmentSatisfaction",]
plt.figure(figsize=(17,10))
w = 1
for i in cols:
    plt.subplot(2,2,w)
    sns.lineplot(y = "PerformanceRating", x = i , data = df)
    plt.title(f"Relationship Between PerformanceRating and {i}", fontsize = 15)
    plt.ylabel("PerformanceRating",fontsize = 13)
    plt.xlabel(i,fontsize = 13)
    w = w+1
plt.tight_layout()
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title("Age Distribution by Attrition Status", fontsize=14)
plt.xlabel("Attrition",fontsize = 12)
plt.ylabel("Age",fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='Gender', hue="PerformanceRating", data=df)
plt.title("Gender Distribution by PerformanceRating Status", fontsize=15)
plt.figure(figsize=(15, 5))
sns.countplot(x='OverTime', hue='EmpWorkLifeBalance', data=df)
plt.title("Overtime Distribution by EmpWorkLifeBalance Status", fontsize=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(y='EmpJobRole', x='YearsWithCurrManager', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='EmpJobRole', hue='BusinessTravelFrequency', data=df)
plt.xticks(rotation = 45)
plt.show()
```

## Multivariate Analysis

```python
selected_columns = [
    'Age', 'EmpJobSatisfaction', 'EmpEnvironmentSatisfaction', 
    'EmpHourlyRate', 'Attrition'
]
sns.pairplot(df[selected_columns], hue='Attrition')
plt.show()
```

```python
travel_attrition = pd.crosstab(df['BusinessTravelFrequency'], df['Attrition'])
travel_attrition.plot(kind='bar', stacked=True, figsize=(8, 6), color=['#1f77b4', '#ff7f0e'])
plt.title("Attrition by Business Travel Frequency")
plt.xlabel("Business Travel Frequency")
plt.ylabel("Number of Employees")
plt.legend(title="Attrition")
plt.show()
```

```python
sns.scatterplot(
    x='TotalWorkExperienceInYears',
    y='ExperienceYearsAtThisCompany',
    hue='Attrition',
    data=df,
)
plt.title("Total Work Experience vs. Experience at Current Company")
plt.show()
```

```python
sns.countplot(x='EmpDepartment', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.title("Attrition by Department")
plt.show()
```
