## Inverse Transforming 

```python
ohe_cols = ['Attrition','Gender']
test_data[ohe_cols] = ohe.inverse_transform(test_data[ohe_cols])
```

```python
le_cols = ['OverTime','EmpJobRole','EmpDepartment','EducationBackground','MaritalStatus']
for i in le_cols:
    test_data[i] = le[i].inverse_transform(test_data[i])
```

```python
ord_cols = ['BusinessTravelFrequency','PerformanceRating']
for col, encoder in ord.items():
    test_data[[col]] = encoder.inverse_transform(test_data[[col]])
```

```python
test_data[["PerformanceRating_original"]] = ord["PerformanceRating"].inverse_transform(target[["PerformanceRating"]])
```
