```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
```

## Univariate Analysis

```python
disc_cols = ["EmpEducationLevel","EmpEnvironmentSatisfaction","EmpJobInvolvement","EmpJobLevel","EmpJobSatisfaction","NumCompaniesWorked",
            "EmpRelationshipSatisfaction","TrainingTimesLastYear","EmpWorkLifeBalance","PerformanceRating"]
len(disc_cols)
```

```python
plt.figure(figsize=(15, 20))
w = 1
for i in disc_cols:
    plt.subplot(5, 2, w)
    sns.barplot(
        x=df[i].value_counts().index,    
        y=df[i].value_counts().values,   
        palette="deep")
    plt.ylabel("Count", fontsize=12)   
    plt.xlabel(i, fontsize=12) 
    plt.title(f"Distribution of {i}", fontsize=14)
    w = w + 1
plt.tight_layout()
```

```python
cont_cols = ['Age','DistanceFromHome','EmpHourlyRate', 'EmpLastSalaryHikePercent','TotalWorkExperienceInYears', 
             'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
len(cont_cols)
```

```python
plt.figure(figsize=(15,15))
w = 1
for i in cont_cols:
    plt.subplot(3,3,w)
    sns.histplot(x = i, data = df, kde= True)
    w = w+1
    plt.title(f"Distribution of {i}", fontsize= 15)
    plt.xlabel(i,fontsize = 12)
plt.tight_layout()
```

## Bivariate Analysis

```python
cols = ["TotalWorkExperienceInYears","ExperienceYearsAtThisCompany", "EmpJobLevel","PerformanceRating",]
plt.figure(figsize=(17,10))
w = 1
for i in cols:
    plt.subplot(2,2,w)
    sns.lineplot(x = "Age", y = i , data = df)
    plt.title(f"Relationship Between Age and {i}", fontsize = 15)
    plt.xlabel("Age",fontsize = 13)
    plt.ylabel(i,fontsize = 13)
    w = w+1
plt.tight_layout()
```

```python
cols = ["EmpLastSalaryHikePercent","EmpEnvironmentSatisfaction",]
plt.figure(figsize=(17,10))
w = 1
for i in cols:
    plt.subplot(2,2,w)
    sns.lineplot(y = "PerformanceRating", x = i , data = df)
    plt.title(f"Relationship Between PerformanceRating and {i}", fontsize = 15)
    plt.ylabel("PerformanceRating",fontsize = 13)
    plt.xlabel(i,fontsize = 13)
    w = w+1
plt.tight_layout()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='Gender', hue="PerformanceRating", data=df)
plt.title("Gender Distribution by PerformanceRating Status", fontsize=15)
plt.figure(figsize=(15, 5))
sns.countplot(x='OverTime', hue='EmpWorkLifeBalance', data=df)
plt.title("Overtime Distribution by EmpWorkLifeBalance Status", fontsize=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='EmpJobRole', hue='BusinessTravelFrequency', data=df)
plt.xticks(rotation = 45)
plt.show()
```

## Multivariate Analysis

```python
selected_columns = [
    'Age', 'EmpJobSatisfaction', 'EmpEnvironmentSatisfaction', 
    'EmpHourlyRate', 'Attrition'
]
sns.pairplot(df[selected_columns], hue='Attrition')
plt.show()
```

```python
sns.scatterplot(
    x='TotalWorkExperienceInYears',
    y='ExperienceYearsAtThisCompany',
    hue='Attrition',
    data=df,
)
plt.title("Total Work Experience vs. Experience at Current Company")
plt.show()
```

```python
sns.countplot(x='EmpDepartment', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.title("Attrition by Department")
plt.show()
```

## Visualizations Based on Performance Ratings 
### (2-Below Average Performance, 3-Average Performance, 4-Above Average Performance)

```python
performance_ratings = [2, 3, 4]

colors = ['salmon', 'lightblue', 'lightgreen']

for rating, color in zip(performance_ratings, colors):
    df_rating = df[df['PerformanceRating'] == rating]
    plt.figure(figsize=(8, 5))
    sns.histplot(df_rating['Age'], bins=20, kde=True, color=color)
    plt.title(f'Age Distribution (Performance Rating = {rating})')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
    
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='EmpJobRole', y='EmpHourlyRate', data=df_rating, palette='Set2')
    plt.xticks(rotation=45)
    plt.title(f'Hourly Rate by Job Role (Performance Rating = {rating})')
    plt.xlabel('Job Role')
    plt.ylabel('Hourly Rate')
    plt.show()
    
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='EmpJobRole', y='EmpJobSatisfaction', data=df_rating, palette='Set1')
    plt.xticks(rotation=45)
    plt.title(f'Job Satisfaction by Job Role (Performance Rating = {rating})')
    plt.xlabel('Job Role')
    plt.ylabel('Job Satisfaction')
    plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='PerformanceRating', y='EmpJobSatisfaction', data=df, palette='muted')
plt.title('Job Satisfaction by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Job Satisfaction')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='PerformanceRating', hue='Attrition', data=df, palette='coolwarm')
plt.title('Attrition Rate by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.show()
```
